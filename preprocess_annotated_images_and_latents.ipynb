```python
# MULTI CLASS STUFF

import numpy as np 
# importing PIL
from PIL import Image
import os 

folder_GIMP = folder + "/eyes_GIMP/"
to_folder = folder + "/eyes_processed/"

do_NIR_as_well = True
os.makedirs(to_folder, exist_ok=True)

for i, id in enumerate(image_ids): 
    # Read image
    mask = Image.open( folder_GIMP + "image_" + str(id) + "_NIR_MASK_ALL.png").convert('L')
    img =  Image.open( folder_GIMP + "image_" + str(id) + ".jpg")

    #mask.show()
    if do_NIR_as_well:
        img_NIR =  Image.open( folder_GIMP + "image_" + str(id) + "_NIR.jpg")
    
    mask = np.array(mask)
    
    print(id)
    #img.show()

    #print(mask.shape)
    print(np.unique(mask))
    print(len(np.unique(mask)))
    # break 
    # backgrounds
    mask[mask < 15] = 0 
    
    mask[mask == 29 ] = 1 #iris
    mask[mask == 76 ] = 2 #pupil
    mask[mask == 150 ] = 3 #sclera
    # mask[mask == 105 ] = 6 # eyelashes 
    
    mask[mask == 177] = 4 # bottom lid whole 
    mask[mask == 174] = 4 # bottom lid whole 

    mask[mask == 121 ] = 5 # bottom lid inner
    mask[mask == 71 ] = 7 # krog iris 
    


    mask[mask == 209 ] = 8 # pupil circle
    mask[mask == 226 ] = 6 # top eyelid
    mask[mask == 222 ] = 6 # top eyelid

    mask[mask == 151 ] = 9 # tear thingie
    mask[mask == 156 ] = 9 # tear thingie
    
    
    mask = Image.fromarray(mask)
    mask.show()
    # break 

    print(np.unique(mask))

    np.save(to_folder + "image_mask" + str(i) + ".npy", mask)


    img.save(to_folder + "image_" + str(i) + ".jpg")
    
    if do_NIR_as_well:
        img_NIR.save(to_folder + "image_" + str(i) + "_NIR.jpg")
```
